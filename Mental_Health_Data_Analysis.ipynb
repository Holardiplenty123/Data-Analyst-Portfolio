# **GROUPING BY ICD CODES TO ANALYSE OTHER VARIABLES**

Finding the relationships between 'ICD Code' and other relevant columns (e.g., Age, Gender, Outcome) to identify potential trends or patterns.


```python
icd_code_counts = df['ICD Code'].value_counts()
plt.figure(figsize=(10, 6))
icd_code_counts.plot(kind='bar')
plt.title('Distribution of ICD Codes')
plt.xlabel('ICD Code')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

I Grouped the DataFrame by 'ICD Code' and calculate descriptive statistics for numerical columns and value counts for categorical columns to explore relationships between 'ICD Code' and other variables and Create box plots for numerical columns and count plots for categorical columns against 'ICD Code' to visualize the relationships.




```python
numerical_cols_for_analysis = ['Age', 'Symptom Severity (1-10)', 'Mood Score (1-10)', 'Sleep Quality (1-10)', 'Physical Activity (hrs/week)', 'Treatment Duration (weeks)', 'Stress Level (1-10)', 'Treatment Progress (1-10)', 'Adherence to Treatment (%)']
categorical_cols_for_analysis = ['Gender', 'Outcome']

for col in numerical_cols_for_analysis:
    plt.figure(figsize=(12, 6))
    sns.boxplot(data=df, x='ICD Code', y=col)
    plt.title(f'Distribution of {col} per ICD Code')
    plt.xlabel('ICD Code')
    plt.ylabel(col)
    plt.show()

for col in categorical_cols_for_analysis:
    plt.figure(figsize=(8, 5))
    sns.countplot(data=df, x='ICD Code', hue=col)
    plt.title(f'Distribution of {col} per ICD Code')
    plt.xlabel('ICD Code')
    plt.ylabel('Count')
    plt.show()
```



# **SUMMARY AND KEY FINDINGS**

*   There are four unique ICD codes in the dataset: 'F32.2', 'F41.0', 'F41.1', and 'F31.1'.
*   'F41.1' is the most frequent ICD code (135 counts), while 'F41.0' is the least frequent (116 counts).
*   Numerical variables such as Age, Symptom Severity, Mood Score, Sleep Quality, Physical Activity, Treatment Duration, Stress Level, Treatment Progress, and Adherence to Treatment show variations in their distributions across different ICD codes, suggesting potential differences in patient characteristics and treatment responses based on diagnosis.
*   There are relationships between ICD codes and categorical variables like Gender and Outcome, indicating that certain diagnoses might be more prevalent in specific genders or have different outcome distributions.

# **GROUPING BY AGE AND OUTCOME**

Grouping the DataFrame by the new age category column and the 'Outcome' column, then count the occurrences in each group to see the distribution of outcomes within each age range.


```python
outcome_distribution_by_age = df.groupby(['Age Group', 'Outcome']).size().unstack(fill_value=0)
print("Distribution of Outcome per Age Group:")
display(outcome_distribution_by_age)
```

```python
outcome_distribution_by_age.plot(kind='bar', stacked=True, figsize=(12, 7))
plt.title('Outcome Distribution by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Outcome')
plt.tight_layout()
plt.show()
```
